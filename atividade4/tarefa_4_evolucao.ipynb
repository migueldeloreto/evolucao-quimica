Adopting a SFR of $5M_☉$ $pc-2$ $Gyr-1$, estimate the fraction of the total stellar mass in the Milky Way that
forms solar twins, meaning stars with masses $m = 1,00±0,05 M_☉$ , in the case of:

A) IMF from Miller & Scalo (1979), given by eq. (5.37) in Maciel’s textbook

```python
import numpy as np
from scipy.integrate import quad

def m_phi_m (m):
    return (3.83/5)*np.exp(-1.09*(np.log10(m)+1.02)**2)

frac_twins= quad(m_phi_m, 0.95, 1.05)
print(frac_twins)
```

B) IMF from Salpeter (1955), $\xi(m) = m \phi(m) = K m^{-1.35}$

```python
def salpeter(m):
    return m**(-1.35)

k=1/quad(salpeter, 0.08, 100)[0]
print(k)

def salpeter(m):
    return k*(m**(-1.35))

frac_twins=quad(salpeter, 0.95, 1.05)
print(frac_twins)
```
